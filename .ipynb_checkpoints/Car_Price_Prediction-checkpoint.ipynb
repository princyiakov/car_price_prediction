```python
df["Levy"] = df["Levy"].replace('-',0)
df['Levy']=df['Levy'].astype(int)
```

```python
df['Engine volume']=df['Engine volume'].astype(float)
df['Cylinders']=df['Cylinders'].astype(float)
df['Airbags']=df['Airbags'].astype(float)
```

```python
df
```

```python
df.drop(["ID"], inplace = True, axis = 1)
```

```python
df['Mileage']=df['Mileage'].astype(int)
```

```python
mileage = df['Mileage'].str.split(" ", n = 1, expand = True) # Split mileage into list of two value 
```

```python
mileage
```

```python
df['Mileage'] = mileage[0]
```

```python
df
```

```python
print(df['Manufacturer'].unique())
print(df['Model'].unique())
print(df['Category'].unique())
print(df['Fuel type'].unique())
print(df['Cylinders'].unique())
print(df['Doors'].unique())
print(df['Drive wheels'].unique())
print(df['Fuel type'].unique())
print(df['Wheel'].unique())
print(df['Color'].unique())
print(df['Engine volume'].unique())


```

```python
eng_vol = df['Engine volume'].str.split(" ", expand = True) # Split mileage into list of two value 
```

```python
eng_vol
```

```python
df['Engine volume'] = eng_vol[0]
df['Turbo'] = eng_vol[1]
df['Turbo'].fillna(0, inplace=True)
df['Turbo'] = df['Turbo'].replace('Turbo',1)
df['Doors']=df['Doors'].map({'04-May':'4_5','02-Mar':'2_3','>5':'5'})
df['Doors']=df['Doors'].astype(str)
```

```python
# Divding the data into Non Categorcial and Categorical value for Encoding 
numeric_df = df.select_dtypes(include=np.number) # Continous
cat_df = df.select_dtypes(include=object) # Categorical

# Initiating encoding of categorical columns
encoding = OrdinalEncoder()
cat_col = cat_df.columns.tolist()
encoding.fit(cat_df[cat_col])
cat_ord = encoding.transform(cat_df[cat_col])
cat_ord = pd.DataFrame(cat_ord, columns= cat_col)
```

```python
cat_ord.reset_index(inplace= True, drop = True)
numeric_df.reset_index(inplace= True, drop = True)
```

```python
final_df=pd.concat([numeric_df, cat_ord],axis=1)
```

```python
# Removing the outliers using IQR Inter Quantile Range)
# Calculate the lower and upper limit by finding the Q1(25%) and Q3(75%) of data and multiplying with 1.5

def find_range(df, col):
    q1, q3 = np.percentile(df[col],25), np.percentile(df[col],75)
    iqr = q3 -q1
    print ("Q3 of the data is {} Q1 os the data is {} and the IQR is {}".format(q1, q3, iqr))
    cut_off = iqr * 1.5
    low, up = q1 - cut_off, q3 + cut_off
    print("The lower limit is {} and upper limit is {}".format(low, up))
    return low, up

    
```

```python
outlier_cols=['Levy','Engine volume','Mileage','Cylinders']
for col in outlier_cols:
    lower,upper=find_range(df,col)
    final_df[col] = final_df[col].clip(lower, upper, axis=0) #clipping the value 
```
